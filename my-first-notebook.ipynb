```python
# download the California list of committees tied to propositions
```

```python
props = pd.read_csv("http://www.firstpythonnotebook.org/_static/committees.csv")
```

```python
props.head() # .head() returns the header and top 5 rows; .info() returns stats
```

```python
props.info()
```

```python
contribs = pd.read_csv("http://www.firstpythonnotebook.org/_static/contributions.csv")
```

```python
contribs.head()
```

```python
contribs.info()
```

```python
# .value_counts() returns a Series of distinct values; 
# .reset_index() converts this series into a DataFrame
```

```python
props.prop_name.value_counts().reset_index()
```

```python
# props[] does what? []is the select-where clause; returned as DataFrame
```

```python
prop = props[props.prop_name == 'PROPOSITION 064- MARIJUANA LEGALIZATION. INITIATIVE STATUTE.']
```

```python
prop.head()
```

```python
prop.info()
```

```python
# merge is akin to SQL join; selecting contribs where prop = prop64
```

```python
merged = pd.merge(prop, contribs, on="calaccess_committee_id")
```

```python
merged.head()
```

```python
merged.info()
```

```python
merged.amount.sum() #doesn't includes contribs under $100
```

```python
merged.committee_position.value_counts()
```

```python
support = merged[merged.committee_position == 'SUPPORT']
```

```python
oppose = merged[merged.committee_position == 'OPPOSE']
```

```python
oppose.amount.sum()
```

```python
support.amount.sum()/merged.amount.sum() #percentage of total large contributions
```

```python
oppose.sort_values("amount", ascending=False).head()
```

```python
merged.groupby("committee_name_x").amount.sum().reset_index().sort_values("amount", ascending=False)
# multi step: group by committee name, calc sum of amount, reformat, sort descending
```

```python
merged.groupby(["contributor_firstname", "contributor_lastname", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False).head()
# multi step: group by fname, lname; calc sum of amount; reformat; sort descending
```

```python
outside = merged[merged.contributor_state != 'CA']
```

```python
outside.amount.sum() /merged.amount.sum() *100
```
